# Financial Analysis of Microsoft, Tesla, and Apple (10-K Data)

### Objective
The objective of this analysis is to extract key financial data from the 10-K filings of Microsoft, Tesla, and Apple for the last three fiscal years and analyze trends using Python. 

The analysis focuses on:
- Total Revenue
- Net Income
- Total Assets
- Total Liabilities
- Operating Cash Flow

The insights derived will demonstrate how structured financial data can support the development of an AI-powered financial chatbot.

### Data Extraction

Financial data was manually extracted from the SEC EDGAR database from the 10-K filings of:

- Microsoft
- Tesla
- Apple

For each company, data for the last three fiscal years (2023–2025) was collected from:
- Consolidated Income Statements
- Consolidated Balance Sheets
- Consolidated Statements of Cash Flows

The extracted data was compiled into an Excel file and converted to CSV format for analysis.

### IMPORT LIBRARIES

```python
import pandas as pd
```

### LOAD DATA

```python
df = pd.read_csv("C:/Users/LENOVO/OneDrive/Desktop/FORAGE-BOC-TASK/Data.csv")
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df = df.dropna(subset=['Year'])
df['Year'] = df['Year'].astype(int)

#df
df.tail(10)
```

```python
df = df.sort_values(by=['Company', 'Year'])
df
```

```python
df['Revenue Growth (%)'] = df.groupby('Company')['Revenue (USD Millions)'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income (USD Millions)'].pct_change() * 100
df['Asset Growth (%)'] = df.groupby('Company')['Assets (USD Millions)'].pct_change() * 100
df['Operating Cash Flow Growth (%)'] = df.groupby('Company')['Operating Cash Flow (USD Millions)'].pct_change() * 100
df.fillna(0, inplace=True)
df
```

### Financial Trend Analysis (2023–2025)
##### Microsoft

Revenue grew consistently in 2024 (15.67%) and 2025 (14.93%).

Net income increased significantly in 2024 (21.80%) and continued growing in 2025 (15.54%).

Assets and operating cash flow also showed strong double-digit growth.

This indicates strong operational efficiency and expanding business performance.

##### Tesla

Revenue growth remained relatively flat, with slight decline in 2025 (-2.93%).

Net income declined sharply in both 2024 (-52.23%) and 2025 (-46.11%).

However, assets continued growing, suggesting continued investment and expansion.

This may indicate margin pressure despite asset growth.

##### Apple

Revenue showed steady growth, especially in 2025 (6.43%).

Net income declined in 2024 but rebounded strongly in 2025 (19.50%).

Asset growth was moderate.

Operating cash flow remained stable overall.

### Methodology

Financial data was manually extracted from each company’s 10-K filings (Microsoft, Tesla, Apple).

Data was compiled into Excel and converted to CSV format.

Data was cleaned and sorted by Company and Year (ascending).

Year-over-year growth was calculated using pandas groupby() and pct_change() functions.

Growth rates were analyzed to identify financial trends across companies.
