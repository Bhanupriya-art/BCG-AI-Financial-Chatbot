# GFC AI Financial Chatbot Prototype  
## Demonstration and Documentation

---

## 1. Project Overview

This project presents a rule-based AI financial chatbot developed to transform structured financial data into interactive insights. The chatbot responds to predefined financial queries using financial data for Microsoft, Tesla, and Apple covering the years 2023–2025.

The objective of this prototype is to demonstrate how financial analysis results can be integrated into a chatbot framework using simple rule-based logic.

---

## 2. How the Chatbot Works

The chatbot operates in four main stages:

### 2.1 Data Loading and Preparation
- The chatbot loads financial data from a CSV file using pandas.
- Completely empty rows are removed.
- Financial columns are converted into numeric format.
- Missing numeric values are handled safely.

### 2.2 Rule-Based Logic
- The chatbot uses predefined `if-else` statements.
- Each supported query is mapped to a specific financial function.
- The user must enter the query exactly as defined.

### 2.3 Financial Data Retrieval
For each recognized query:
- The chatbot filters relevant company data.
- Performs necessary calculations (e.g., averages, yearly changes).
- Returns a structured, readable financial insight.

### 2.4 Error Handling
If a query is not recognized:
- The chatbot returns a message explaining that only predefined queries are supported.
- It suggests valid queries the user can ask.

---

## 3. Supported Predefined Queries

The chatbot currently supports the following exact queries:

1. What is the total revenue of Microsoft?
2. How has Apple's net income changed over the last year?
3. Which company has the highest average revenue?
4. Compare Tesla and Microsoft revenue.
5. What was Apple's net income in 2023?

## 5. Limitations

- The chatbot only supports exact predefined queries.
- No natural language processing (NLP) flexibility.
- No machine learning integration.
- No real-time financial data updates.
- Command-line interface only (no web interface).

```python
import pandas as pd

df = pd.read_csv('Data.csv')
df.dropna(how='all', inplace=True)

numeric_cols = [
    "Revenue (USD Millions)",
    "Net Income (USD Millions)",
    "Assets (USD Millions)",
    "Liabilities (USD Millions)",
    "Operating Cash Flow (USD Millions)"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df[numeric_cols] = df[numeric_cols].fillna(0)


def total_revenue(company):
    data = df[df["Company"] == company]
    total = data["Revenue (USD Millions)"].sum()
    return f"The total revenue of {company} is {total} million USD."

def net_income_change(company):
    data = df[df["Company"] == company].sort_values("Year")
    
    if len(data) >= 2:
        last_year = data.iloc[-1]["Net Income (USD Millions)"]
        previous_year = data.iloc[-2]["Net Income (USD Millions)"]
        change = last_year - previous_year
        
        if change > 0:
            trend = "increased"
        elif change < 0:
            trend = "decreased"
        else:
            trend = "remained the same"
        
        return f"{company}'s net income has {trend} by {abs(change)} million USD over the last year."
    else:
        return "Not enough data available to calculate net income change."

def highest_average_revenue():
    summary = df.groupby("Company")["Revenue (USD Millions)"].mean()
    company = summary.idxmax()
    return f"{company} has the highest average revenue over the selected years."

def compare_revenue(company1, company2):
    avg1 = df[df["Company"] == company1]["Revenue (USD Millions)"].mean()
    avg2 = df[df["Company"] == company2]["Revenue (USD Millions)"].mean()
    
    return (
        f"Average Revenue:\n"
        f"{company1}: {avg1:.2f} million USD\n"
        f"{company2}: {avg2:.2f} million USD"
    )

def net_income_year(company, year):
    data = df[(df["Company"] == company) & (df["Year"] == year)]
    
    if not data.empty:
        value = data["Net Income (USD Millions)"].values[0]
        return f"{company}'s net income in {year} was {value} million USD."
    else:
        return "Data not found for that year."


def simple_chatbot(user_query):
    user_query = user_query.lower()

    if user_query == "what is the total revenue of microsoft?":
        return total_revenue("Microsoft")

    elif user_query == "how has apple's net income changed over the last year?":
        return net_income_change("Apple")

    elif user_query == "which company has the highest average revenue?":
        return highest_average_revenue()

    elif user_query == "compare tesla and microsoft revenue.":
        return compare_revenue("Tesla", "Microsoft")

    elif user_query == "what was apple's net income in 2023?":
        return net_income_year("Apple", 2023)

    else:
        return (
            "Sorry, I can only respond to predefined queries.\n"
            "Try asking:\n"
            "- What is the total revenue of Microsoft?\n"
            "- How has Apple's net income changed over the last year?\n"
            "- Which company has the highest average revenue?\n"
            "- Compare Tesla and Microsoft revenue.\n"
            "- What was Apple's net income in 2023?"
        )

print("GFC Financial Chatbot Prototype Ready!")
print("Type your query exactly as shown or type 'exit' to quit.\n")

while True:
    user_input = input("You: ")
    
    if user_input.lower() == "exit":
        print("Chatbot session ended.")
        break
    
    response = simple_chatbot(user_input)
    print("Bot:", response)
```
